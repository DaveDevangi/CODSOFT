TASK 1: CHATBOT WITH RULE-BASED RESPONSES

Build a simple chatbot that responds to user inputs based on
predefined rules. Use if-else statements or pattern matching
techniques to identify user queries and provide appropriate
responses. This will give you a basic understanding of natural
language processing and conversation flow.

```python
import random

def chatbot_response(user_input):
    user_input = user_input.lower()

    greetings = ["Hey there! How can I brighten your day?", "Hello! What's on your mind today?", "Hi! How can I assist you?"]
    feelings = ["I'm functioning optimally, thanks for asking! What about you?", "I'm all circuits and signals, ready to help you!", "I'm just a friendly AI, ready to assist! How are you?"]
    farewells = ["Take care! It was great chatting with you!", "Goodbye! Don't hesitate to reach out again.", "Farewell! Hope to see you soon!"]
    names = ["I'm your digital companion, always here to help!", "You can call me your trusty AI assistant.", "I'm known as your AI sidekick."]
    helps = ["Absolutely! What do you need help with?", "Sure thing! What can I do for you?", "Of course! Let's solve your problem together."]
    unsure = ["That's an interesting one. Could you elaborate a bit?", "Hmm, I'm not sure I understand. Can you rephrase?", "I might need a bit more info to give you a proper response."]

    if 'hello' in user_input:
        return random.choice(greetings)
    elif 'how are you' in user_input:
        return random.choice(feelings)
    elif 'bye' in user_input:
        return random.choice(farewells)
    elif 'name' in user_input:
        return random.choice(names)
    elif 'help' in user_input:
        return random.choice(helps)
    else:
        return random.choice(unsure)

while True:
    user_input = input("You: ")
    if 'bye' in user_input.lower():
        print("Bot:", random.choice(["Take care! It was great chatting with you!", "Goodbye! Don't hesitate to reach out again.", "Farewell! Hope to see you soon!"]))
        break
    response = chatbot_response(user_input)
    print("Bot:", response)
```
